```python
import matplotlib.pyplot as plt
import numpy as np
import csv
from matplotlib.pyplot import figure
import seaborn as sns
import pandas as pd
from scipy.integrate import simps
from numpy import trapz
```

```python
u1 = []
v1 = []
u2 = []
v2 = []
u3 = []
v3 = []
u4 = []
v4 = []

data = 'U_V_data.csv'
with open(data, 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    for row in csvreader:
      if row[0]=='U_1':
        pass
      else:
        u1.append(float(row[0]))
        v1.append(float(row[1]))
        u2.append(float(row[2]))
        v2.append(float(row[3]))
        u3.append(float(row[4]))
        v3.append(float(row[5]))
        u4.append(float(row[6]))
        v4.append(float(row[7]))
```

STANDARD DEVIATION OF THE FLUCTUATING VELOCITIES

```python
N=len(t)
```

```python

mean_u1 = np.mean(u1)
mean_v1 = np.mean(v1)
mean_u2 = np.mean(u2)
mean_v2 = np.mean(v2)
mean_u3 = np.mean(u3)
mean_v3 = np.mean(v3)
mean_u4 = np.mean(u4)
mean_v4 = np.mean(v4)

u1_prime = [u-mean_u1 for u in u1]
u2_prime = [u-mean_u2 for u in u2]
u3_prime = [u-mean_u3 for u in u3]
u4_prime = [u-mean_u4 for u in u4]
v1_prime = [u-mean_v1 for u in v1]
v2_prime = [u-mean_v2 for u in v2]
v3_prime = [u-mean_v3 for u in v3]
v4_prime = [u-mean_v4 for u in v4]
```

```python
a1=0
a2=0
a3=0
a4=0
a5=0
a6=0
a7=0
a8=0

for i in range(N):
  a1 = a1 + u1_prime[i]**2
  a2 = a2 + v1_prime[i]**2
  a3 = a3 + u2_prime[i]**2
  a4 = a4 + v2_prime[i]**2
  a5 = a5 + u3_prime[i]**2
  a6 = a6 + v3_prime[i]**2
  a7 = a7 + u4_prime[i]**2
  a8 = a8 + v4_prime[i]**2

sig_u1 = np.sqrt(a1/N)
sig_v1 = np.sqrt(a2/N)
sig_u2 = np.sqrt(a3/N)
sig_v2 = np.sqrt(a4/N)
sig_u3 = np.sqrt(a5/N)
sig_v3 = np.sqrt(a6/N)
sig_u4 = np.sqrt(a7/N)
sig_v4 = np.sqrt(a8/N)
```

SKEWNESS

```python
a1=0
a2=0
a3=0
a4=0
a5=0
a6=0
a7=0
a8=0

for i in range(N):
  a1 = a1 + u1_prime[i]**3
  a2 = a2 + v1_prime[i]**3
  a3 = a3 + u2_prime[i]**3
  a4 = a4 + v2_prime[i]**3
  a5 = a5 + u3_prime[i]**3
  a6 = a6 + v3_prime[i]**3
  a7 = a7 + u4_prime[i]**3
  a8 = a8 + v4_prime[i]**3

skew_u1 = a1/N
skew_v1 = a2/N
skew_u2 = a3/N
skew_v2 = a4/N
skew_u3 = a5/N
skew_v3 = a6/N
skew_u4 = a7/N
skew_v4 = a8/N
```

KURTOSIS

```python
a1=0
a2=0
a3=0
a4=0
a5=0
a6=0
a7=0
a8=0

for i in range(N):
  a1 = a1 + u1_prime[i]**4
  a2 = a2 + v1_prime[i]**4
  a3 = a3 + u2_prime[i]**4
  a4 = a4 + v2_prime[i]**4
  a5 = a5 + u3_prime[i]**4
  a6 = a6 + v3_prime[i]**4
  a7 = a7 + u4_prime[i]**4
  a8 = a8 + v4_prime[i]**4

kurt_u1 = a1/N
kurt_v1 = a2/N
kurt_u2 = a3/N
kurt_v2 = a4/N
kurt_u3 = a5/N
kurt_v3 = a6/N
kurt_u4 = a7/N
kurt_v4 = a8/N
```

```python
sigma_u = [sig_u1,sig_u2,sig_u3,sig_u4]
sigma_v = [sig_v1,sig_v2,sig_v3,sig_v4]

skewness_u = [skew_u1, skew_u2, skew_u3, skew_u4]
skewness_v = [skew_v1, skew_v2, skew_v3, skew_v4]

kurtosis_u = [kurt_u1, kurt_u2, kurt_u3, kurt_u4]
kurtosis_v = [kurt_v1, kurt_v2, kurt_v3, kurt_v4]
```

```python
def analyze(R,tt,limit):
  Q_u1 = []
  Q_t1 = []

  w_u1 = []
  w_t1 = []

  for i in range(len(R)):
    if abs(R[i]) > limit:
      Q_u1.append(R[i])
      Q_t1.append(tt[i])
    else:
      w_u1.append(R[i])
      w_t1.append(tt[i])
  figure(figsize=(16, 6), dpi=100)
  plt.scatter(Q_t1,Q_u1)
  plt.scatter(w_t1,w_u1)
  return 
```

```python
def timeAvg(t):
  areas = []
  for listo in t:
    y=np.array(listo)
    #print(y)
    #area = simps(y, dx=0.0033)
    area = abs(trapz(y, dx=0.0033))
    if area > 0.00000001:
      areas.append(area)
    else:
      pass
  print(areas)
  return np.mean(areas)

def timeScale(u):
  array1 = []
  array2 = []
  check = True
  for i in range(len(u)):
    if u[i]>0 and check == True:
      array2.append(u[i])
    elif u[i]>0 and check == False:
      check = True
      array2.append(u[i])
    elif u[i]<=0 and check == True:
      if i>0:
        check = False
        if len(array2)>3:
          array1.append(array2)
          array2 = []
        else:
          pass
      else:
        check=False
    else:
      pass
  avg_t = timeAvg(array1)
  return avg_t
```

```python
#finding the time and length scales
t_avg_U1=timeScale(R_U1)
t_avg_U2=timeScale(R_U2)
t_avg_U3=timeScale(R_U3)
t_avg_U4=timeScale(R_U4[:2000])
t_avg_V1=timeScale(R_V1)
t_avg_V2=timeScale(R_V2)
t_avg_V3=timeScale(R_V3)
t_avg_V4=timeScale(R_V4[:2000])

L_avg_U1 = mean_u1*t_avg_U1
L_avg_U2 = mean_u1*t_avg_U2
L_avg_U3 = mean_u1*t_avg_U3
L_avg_U4 = mean_u1*t_avg_U4
L_avg_V1 = mean_u1*t_avg_V1
L_avg_V2 = mean_u1*t_avg_V2
L_avg_V3 = mean_u1*t_avg_V3
L_avg_V4 = mean_u1*t_avg_V4
```

```python
def mikes_pdf(y):
  bins=50
  b = list(np.linspace(min(y),max(y),bins))
  pdf_y = []
  histogram_y = []
  for i in range(bins-1):
    count = 0
    for j in range(len(y)):
      if y[j] >= b[i] and y[j] < b[i+1]:
        count = count+1
    probobility = count/len(y)
    histogram_y.append(count)
    pdf_y.append(probobility)
  x = b[0:(len(b)-1)]
  return x, pdf_y, histogram_y
```
